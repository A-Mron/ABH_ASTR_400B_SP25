# Part A

Create a function that returns the SFR for a given luminosity (NUV, FUV, TIR, Halpha)

$Log( {\rm SFR} (M_\odot/year)) = Log(Lx (erg/s)) - Log(Cx)$ 

Including corrections for dust absorption 

Kennicutt & Evans 2012 ARA&A Equation 12 and Table 1, 2

```python
def StarFormarionRate(L, Type, TIR=0):
    '''
    Function that compites the star formation rate of a galaxy following  Kennicutt and Evans 2012 
    Eq 12 (ARA&A 50)
    Paramerters:
        L: 'float', luminocity of the galaxy [erg/s]
        Type: 'str', Wavelength: 'FUV', 'NUV', 'TIR', 'Halpha'
        TIR: 'float', Total infaref lumnocity [erg/s] (default=0)
    Output:
        SFR: 'float', Log of Stat formation rate [Msun/yr]
    '''
    if (Type == 'FUV'):
        logCx = 43.35 #Calibration form LFUV to SFR
        TIRc = 0.46 #correction for dust absorbtion
    elif (Type == 'NUV'):
        logCx = 43.17
        TIRc = 0.27
    elif (Type == 'Halpha'):
        logCx = 41.27
        TIRc = 0.0024
    elif (Type == 'TIR'):
        logCx = 43.41
        TIRc = 0
    else:
        print("WARNING: Missing Wavelength: Accepting FUV, NUV, Halpha, or TIR")
    
    #correct the Lum for dust using the TIR
    Lcorr = L + TIRc*TIR

    #star formation rate
    SFR = np.log10(Lcorr) - logCx

    return SFR
```

# Part B Star formation main sequence

1) Write a function that returns the average SFR of a galaxy at a given redshift, given its stellar mass

2) What is the average SFR of a MW mass galaxy today? at z=1?

3) Plot the SFR main sequence for a few different redshifts from 1e9 to 1e12 Msun.


From Whitaker 2012:

log(SFR) = $\alpha(z)({\rm log}M_\ast - 10.5) + \beta(z)$

$\alpha(z) = 0.7 - 0.13z$

$\beta(z) = 0.38 + 1.14z - 0.19z^2$

# Step 1

```python
def StarFormMainSequ(Mstar,z):
    '''
    Function that computes the aberage SFR of a galaxy as a function of stellar mass and redshift
    Parameters:
        Mstar: 'float', Sellar mass of the galaxy [Msun]
        z: 'float', redshift [unitless]
    Outputs:
        SFRMS: 'float', Log of SFR [Msun/yr]
    '''
    alpha = 0.7 -0.13*z
    beta = 0.38 + 1.14*z - 0.19*z**2

    SFRMS = alpha*(np.log10(Mstar) - 10.5) + beta
    return SFRMS
```

# Step 2

```python
# MW at z=0
MWmstar = 7.5e10
```

```python
# MW at z = 1
print(StarFormMainSequ(MWmstar,0))
print(StarFormMainSequ(MWmstar,1))
```

# Step 3

```python
# create an array of stellar masses
Mass = np.linspace(1e8,1e12)
```

```python

fig = plt.figure(figsize=(8,8), dpi=500)
ax = plt.subplot(111)

# add log log plots
plt.plot(np.log10(Mass), StarFormMainSequ(Mass,0),color = 'r', linewidth = 3, label='z = 0')
plt.plot(np.log10(Mass), StarFormMainSequ(Mass,1),color = 'b', linewidth = 3, label='z = 1')
plt.plot(np.log10(Mass), StarFormMainSequ(Mass,2),color = 'k', linewidth = 3, label='z = 2')
plt.plot(np.log10(Mass), StarFormMainSequ(Mass,3),color = 'g', linewidth = 3, label='z = 3')
# Add axis labels
plt.xlabel('Log(Mstar (M$_\odot$))', fontsize=12)
plt.ylabel('Log(SFR (M$_\odot$/year))', fontsize=12)


#adjust tick label font size
label_size = 12
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')


# Save file
plt.savefig('Lab8_SFR_MainSequence.png')
```
